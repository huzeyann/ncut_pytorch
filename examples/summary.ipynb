```python
# %%
from pathlib import Path

import numpy as np
import torch
from IPython.display import display
from PIL import Image

from ncut_pytorch.predictor import NcutDinov3Predictor
from ncut_pytorch.utils.sample import FPS_DIMENSION_REDUCTION_METHOD

REPO_ROOT = Path.cwd()
if not (REPO_ROOT / "ncut_pytorch").exists():
    REPO_ROOT = Path("/Users/bytedance/projects/ncut_pytorch")

IMAGE_DIR = REPO_ROOT / "examples" / "images"
OUTPUT_DIR = REPO_ROOT / "examples" / "outputs"
OUTPUT_DIR.mkdir(exist_ok=True)

MODEL_CFG = "dinov3_vitb16"
INPUT_SIZE = (512, 512)
BATCH_SIZE = 1
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEGMENTS = [10, 25, 50, 100]
SEED = 0

torch.manual_seed(SEED)
np.random.seed(SEED)

images = [Image.open(IMAGE_DIR / f"view_{i}.jpg").convert("RGB") for i in range(4)]
predictor = NcutDinov3Predictor(
    model_cfg=MODEL_CFG,
    input_size=INPUT_SIZE,
    batch_size=BATCH_SIZE,
).to(DEVICE)
predictor.set_images(images)
summary_image = predictor.summary(n_segments=SEGMENTS, draw_border=True)
output_path = OUTPUT_DIR / f"summary_{MODEL_CFG}_{INPUT_SIZE[0]}_{FPS_DIMENSION_REDUCTION_METHOD}.png"
summary_image.save(output_path)
output_path
```
